# Задание 3

1. Оценить погрешность в определении корней уравнения $a y^3+d=0$, если величины $a=1$ и $d=8$ заданы с точностью $\Delta(a)=10^{-3}$ и $\Delta(d)=10^{-3}$.

Решение:

Назовём $P(a, d, y) = a y^3 + d$. При $y_0 = -2 :$ $P(a, d, y_0) = 0$. Пусть $\Delta(y_0) : P(a + \Delta(a), d + \Delta(d), y + \Delta(y_0)) = 0$, считаем $\Delta(a) \ll a, \Delta(d) \ll d$, найдём $|\Delta(y_0)|$:

$$
0 = \Delta(P) \approx \Delta(a) y_0^3 + 3 y_0^2 \Delta(y_0) + \Delta(d)  
$$

$$
|\Delta(y_0)| \leq \frac{|\Delta(d)| + |\Delta(a)y_0^3|}{|3 y_0^2 a|} = 7,5 \cdot 10^{-4}
$$

2. Определить оптимальный шаг численного дифференцирования $h_{\text{opt}}$ при использовании для вычисления производной приближенной формулы

$$
u^{\prime}(x) \approx \frac{u(x-2 h)-8 u(x-h)+8 u(x+h)-u(x+2 h)}{12 h},
$$

Какой порядок аппроксимации имеет эта формула? Известно, что, $\left|u^{(5)}(t)\right| \leq M_5$, а значения функций вычисляются с точностью $\Delta u$.

Решение:

Запишем $u(x - 2h)$, $u(x - h)$, $u(x + h)$ и $u(x + 2h)$ в ряды Тейлора с остаточными членами пятого порядка в форме Лагранжа и найдём разность между реальным $u'(x)$ и предложенной формулой:

$$
u^{\prime}(x) - \frac{u(x-2 h)-8 u(x-h)+8 u(x+h)-u(x+2 h)}{12 h} = \frac{h^4}{180}(u^{(5)}(t_3)+u^{(5)}(t_2)-4u^{(5)}(t_1)-4u^{(5)}(t_4))
$$

Из этой разности очевиден порядок аппроксимации $O(h^4)$.

Также максимальный шаг $h_{opt}$, при котором погрешность $u'$ будет определяться погрешностью $\Delta(u)$:

$$
20\Delta(u)/12h_{opt} \geq \frac{h_{opt}^4}{180}(M_5 + M_5 + 4M_5 + 4M_5)
$$

$$
h_{opt} = (360 \frac{\Delta(u)}{M_5})^{1/3}
$$


# Задание 4

С каким числом верных знаков (или относительной погрешностью) должен быть известен свободный член в уравнении $x^2-2 x+0.999993751=0$, чтобы корни имели четыре верных знака?

Решение:

Пусть $z = 0.999993751$.

$x_1, x_2 = 1 \pm \sqrt{1 - z}$. Хотим $\Delta(x) = 10^{-4}$

$$
\Delta(x) \approx \frac{\Delta(z)}{2\sqrt{1 - z}}
$$

$$
\Delta(z) = 2 \Delta(x) \sqrt{1-z} = 5 \cdot 10^{-7} \approx 10^{-8}
$$

Таким образом, нужно знать свободный член с восмью знаками после запятой

1. Сгенерируйте случайную матрицу $N \times N$, в которой каждый элемент взят из гауссова нормального распределения с фиксированными параметрами. Сделайте из этой матрицы симметричную, добавив к ней ее транспонированную. Постройте гистограмму собственных значений получившейся матрицы (подберите достаточно большое $N$ и удачное количество бинов). Повторите эксперимент $M$ раз ( $M \sim 50$, здесь цикл разрешается) и отобразите разброс высоты каждого бина гистограммы с помощью креста ошибок.

```python
N = 400
M = 100
a = 0.0
b = 1.0
bins = 200
```

```python
def generate_symmetric_matrix(N, loc=0.0, scale=1.0):
    A = np.random.normal(loc, scale, size=(N, N))
    return A + A.T
```

```python
eigenvalues = np.zeros((M, N))
```

```python
for i in range(M):
    eigenvalues[i] = np.linalg.eigvals(generate_symmetric_matrix(N, a, b))
```

```python
eigenvalues = np.sort(eigenvalues)
print(eigenvalues)
```

```python
bin_heights_all, bin_edges_all, _ = plt.hist(np.transpose(eigenvalues), bins=bins, alpha=0.5)
#print(bin_heights_all)
bin_centers_all = 0.5 * (bin_edges_all[1:] + bin_edges_all[:-1])
#bin_centers_med = np.mean(bin_centers_all, axis=1)
#bin_centers_std = np.std(bin_centers_all, axis=1)
bin_heights_med = np.mean(bin_heights_all, axis=0)
bin_heights_std = np.std(bin_heights_all, axis=0)
```

```python
plt.title("Гистограмма собственных значений")
plt.xlabel("Значение")
plt.ylabel("Частота")
plt.errorbar(bin_centers_all, bin_heights_med, yerr=bin_heights_std, fmt='k+')
plt.show()
```
